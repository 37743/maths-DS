```python
import numpy as  np
import matplotlib.pyplot as plt
```

```python
x = np.array([5,6,8,9,8])
y = np.array([6,7,2,1,3])

corr_mat = np.corrcoef(x,y)
correlation = corr_mat[1,0]
print(correlation)
print("Positive") if correlation > 0 else print("Negative")
```

```python
m, b = np.polyfit(x,y,1)
plt.scatter(x, y, color='purple', label = 'Data Points')
plt.plot(x, m*x+b, color='orange', label='Line of Best Fit')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("Linear Correlation Visualization of X and Y")
plt.legend()
plt.show()
```

```python
import scipy.stats as st
```

```python
sample_mean = 90
population_mean = 82
pop_std = 20
sample_size = 81
alpha = 0.05

z_score = (sample_mean-population_mean)/(pop_std/np.sqrt(sample_size))
z_critical = st.norm.ppf(1-alpha) # z_alpha
p_value = 1 - st.norm.cdf(z_score)
print('Z-score:',z_score)
print('Z-critical:',z_critical)

if z_score > z_critical:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")

if p_value < alpha:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```
